### Generate csv for sentiment analysis

```python
import os
import pandas as pd
```

```python
pathname_10k = "/Users/kiranrawat/Desktop/portfolio-risk-assessment-using-ml/data/scraped_10k"
pathname_10q = "/Users/kiranrawat/Desktop/portfolio-risk-assessment-using-ml/data/scraped_10q"
cik_tickers = '/Users/kiranrawat/Desktop/portfolio-risk-assessment-using-ml/AllSecTickers.csv'
```

#### generate cik-mapping csv for  10-K SEC filings

```python
# cik_tickers = '/Users/kiranrawat/Desktop/portfolio-risk-assessment-using-ml/AllSecTickers.csv'
cikdir = '/Users/kiranrawat/Desktop/portfolio-risk-assessment-using-ml/data/ciklist/10-K'

cikListFile = pd.read_csv(cik_tickers)

os.chdir(pathname_10k)
dir_list = [fname for fname in os.listdir()if 
                 (fname.startswith('0'))]

#filings_table = filings_table[(filings_table['Filing Date'] >= '2008-01-01')]

for dirs in dir_list:
    file_list = [clean_data for clean_data in os.listdir(dirs+'/clean_data')] 
    #remove .ds_store file if any
    if '.DS_Store' in file_list:
        file_list.remove('.DS_Store') 
    dates = [x[-14:-4] for x in file_list]
    data = pd.DataFrame(dates,columns=['Date'])
    data['sec_type']='10-K'
    data['filename'] = [x for x in file_list]
    data['cik'] = data['filename'].str.split("_", n = 0).str[0].astype('int64')
    result = pd.merge(data, cikListFile, on='cik')
    result = result[(result.Date >= '2008-01-01')]
    output_filename = (data['cik'][0]).astype('str')
    result.to_csv(cikdir + '/' + output_filename+'.csv', mode = 'a', index=False)
```
